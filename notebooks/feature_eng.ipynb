## Feature Engineering

Features that we want:
- count transactions for the time window DONE
- count transactions in the past 3 time windows DONE
- count transactions in the past 6 time windows DONE
- trend of transactions in the past 3 time windows (goes up or down this window compared to the previous window avg) DONE
- trend of transactions in the past 6 time windows (goes up or down this window compared to the previous window avg) DONE
- total transactions this window DONE
- total transactions in the past 3 time windows DONE
- total transactions in the past 6 time windows DONE
- category on tiers based on amount transacted
- active days, weeks and months in this time window DONE
- active days, weeks and months in the past 3 time windows DONE
- active days, weeks and months in the past 6 time windows DONE
- day since first activity  DONE
- amount of unique contracts this time window DONE
- amount of unique contracts in the past 3 time windows DONE
- amount of unique contracts in the past 6 time windows DONE
- days since last transaction DONE

What we need:
- Complete the dataset with the missing months for a user
- Define churn or set different tiers of usage

```python
def award_transaction_points(x): #1,1000,10000,50000,250000,500000,1000000
    if x < 0.00001:
        return 0
    elif x < 1000: 
        return 1
    elif x < 10000:
        return 2
    elif x < 50000:
        return 3
    elif x < 250000:
        return 4
    elif x < 500000:
        return 5
    elif x < 1000000:
        return 6

def award_transaction_count_points(x): # 5,10, 25,50,100
    if x < 0.00001:
        return 0
    elif x < 1: 
        return 1
    elif x < 2:
        return 2
    elif x < 3:
        return 3
    elif x < 4:
        return 4
    elif x < 100:
        return 5
    else:
        return 6
```

```python
def transform_data(df):
    df_ = df.copy()
    df_['Active Months'] = 1 #filling the deleted column
    # Define today's date rounded to the start of the month
    df_["month_start_"] = pd.to_datetime(df_['month_start']).dt.tz_localize(None)

    today = pd.Timestamp.today().normalize() - MonthBegin(1)
    today = today.tz_localize(None)

    def resample_and_fill(group):
        # Resample and fill dates within the existing range
        resampled = group.set_index('month_start_').resample('MS').first()

        # Create a date range from the min date in the group to today
        full_range = pd.date_range(start=group['month_start_'].min(), end=today, freq='MS')

        # Reindex the resampled DataFrame to the full date range
        return resampled.reindex(full_range)

    # Apply the function to each group and reset the index
    df_ = df_.groupby('User Address').apply(resample_and_fill).drop('User Address', axis=1).reset_index(level=0).reset_index()
    df_["month_start_"] = df_["index"]

    df_ = df_.sort_values(by=['month_start_','User Address'])
    #replace for each user the ["Initial Active Date"] with the first one7
    df_["Last Active Date"] = df_.groupby('User Address')["Initial Active Date"].fillna(method = "ffill")
    df_["Initial Active Date"] = df_.groupby("User Address")["Initial Active Date"].transform("first")
    df_ = df_.fillna(0)

    # Convert 'month_start' to datetime if it's not already
    df_['month_start'] = pd.to_datetime(df_['month_start_'])

    # Calculate rolling sums for past 3 and 6 time windows
    for col in ['Transactions Count', 'Amount of total transactions ($)','Interacted TX Contracts','Active Days','Active Months','Active Weeks']:
        df_.sort_values(by=['User Address', 'month_start'], inplace=True)  # Ensure data is sorted

        # Use groupby and rolling with a time-based offset
        df_[f'{col} - past 3 windows'] = df_.groupby('User Address')[col].shift(1).rolling(window=3).mean().reset_index(level=0, drop=True).fillna(0)
        df_[f'{col} - past 6 windows'] = df_.groupby('User Address')[col].shift(1).rolling(window=6).mean().reset_index(level=0, drop=True).fillna(0)

    # Transaction trends
    df_['Trend Transactions - past 3 windows'] = np.where(df_['Transactions Count - past 3 windows'] == 0, 0 ,(df_['Transactions Count'] - df_['Transactions Count - past 3 windows'] )/ (df_['Transactions Count - past 3 windows'] ))
    df_['Trend Transactions - past 6 windows'] = np.where(df_['Transactions Count - past 6 windows'] == 0, 0 ,(df_['Transactions Count'] - df_['Transactions Count - past 6 windows'] )/ (df_['Transactions Count - past 6 windows'] ))

    df_['Last Active Date'] = pd.to_datetime(df_['Last Active Date']).dt.tz_localize(None)
    df_['Initial Active Date'] = pd.to_datetime(df_['Initial Active Date']).dt.tz_localize(None)
    df_['month_start'] = pd.to_datetime(df_['month_start'])

    df_['Days since last transaction'] = (df_['month_start'] + pd.DateOffset(months=1) - df_['Last Active Date']).dt.days
    df_['Days since first activity'] = (df_['month_start'] + pd.DateOffset(months=1) - df_['Initial Active Date']).dt.days

    df_['Transaction Points Dollars'] = df_['Amount of total transactions ($)'].apply(lambda x: award_transaction_points(x))
    df_['Transaction Points Count'] = df_["Transactions Count"].apply(lambda x: award_transaction_points(x))

    #objective variables
    df_["Objetive_next_month"] = df_.groupby("User Address")["Transactions Count"].shift(-1)
    df_["Objetive_next_3_months"] = df_.groupby("User Address")[f'Transactions Count'].rolling(window=3).mean().groupby("User Address").shift(-3).reset_index(level=0, drop=True)
    df_["Objetive_next_6_months"] = df_.groupby("User Address")[f'Transactions Count'].rolling(window=6).mean().groupby("User Address").shift(-6).reset_index(level=0, drop=True)

    df_['Trend_Objetive_next_month'] = np.where(df_['Objetive_next_month'].isna(), np.NaN, np.where(df_['Transactions Count'] == 0, 0 ,(df_['Objetive_next_month'] - df_['Transactions Count'] )/ (df_['Transactions Count'] ))).clip(-1,1)
    df_['Trend_Objetive_next_3_months'] = np.where(df_['Objetive_next_3_months'].isna(), np.NaN,np.where(df_['Transactions Count'] == 0, 0 ,(df_['Objetive_next_3_months'] - df_['Transactions Count'] )/ (df_['Transactions Count'] ))).clip(-1,1)
    df_['Trend_Objetive_next_6_months'] = np.where(df_['Objetive_next_6_months'].isna(), np.NaN,np.where(df_['Transactions Count'] == 0, 0 ,(df_['Objetive_next_6_months'] - df_['Transactions Count'] )/ (df_['Transactions Count'] ))).clip(-1,1)
    
    colums = ['month_start','User Address', 'Active Days', 'Active Months', 'Active Weeks','Amount of total transactions ($)', 
       'Days since first activity', 'Days since last transaction', 'Interacted TX Contracts', 'Transactions Count',
       'Transactions Count - past 3 windows','Transactions Count - past 6 windows',
       'Amount of total transactions ($) - past 3 windows','Amount of total transactions ($) - past 6 windows','Interacted TX Contracts - past 3 windows',
       'Interacted TX Contracts - past 6 windows','Active Days - past 3 windows', 'Active Days - past 6 windows',
       'Active Months - past 3 windows', 'Active Months - past 6 windows','Active Weeks - past 3 windows', 'Active Weeks - past 6 windows',
       'Trend Transactions - past 3 windows','Trend Transactions - past 6 windows', 'Trend_Objetive_next_3_months',
       'Transaction Points Dollars','Transaction Points Count','Trend_Objetive_next_month'] #'Objetive_next_month', 'Objetive_next_3_months', 'Objetive_next_6_months','Trend_Objetive_next_month', 'Trend_Objetive_next_6_months',
       # once we know if we are going for the 1, 3 or 6 months, we can drop the other objetives
    return df_[colums]
```

```python
df_layer_zero_ = transform_data(df_layer_zero)
df_zksync_ = transform_data(df_zksync)
df_base_ = transform_data(df_base)
df_arbitrum_ = transform_data(df_arbitrum)
df_gnosis_ = transform_data(df_gnosis)
df_celo_ = transform_data(df_celo)
df_polygon_ = transform_data(df_polygon)
df_optimism_ = transform_data(df_optimism)
```
